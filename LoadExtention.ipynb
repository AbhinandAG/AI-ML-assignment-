```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("loadextension.csv")
print(data)
```

```python
x=data[["Load (N)"]]
y=data[["Extension (mm)"]]

plt.scatter(x,y)
plt.xlabel("Load (N)")
plt.ylabel("Extension (mm)")
plt.title("Load VS Extension")
plt.show()
```

```python
model=LinearRegression()
model.fit(x,y)
print("Coefficient:",model.coef_)
print("Intercept:",model.intercept_)
```

```python
value=int(input("Enter the load:"))
print("Extension:",model.predict([[value]]))
```
